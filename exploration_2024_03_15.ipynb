```python
import polars as pl
import time
from pathlib import Path

# ================= 配置区域 =================
# 确保这里的路径指向你刚才生成的 merged_all.parquet
BASE_DIR = Path("/Users/wanganbo/Desktop/new")
MERGED_FILE = BASE_DIR / "final_joined_data/trades_and_quotes_joined.parquet"

K_FWD_TRADES = 10  # MR (均值回归) 向前看10笔交易
# ===========================================

def get_microstructure_stats(merged_file_path):
    print(f"🚀 开始分析文件: {merged_file_path}")

    # 1. 扫描文件并获取所有股票列表
    lf_all = pl.scan_parquet(str(merged_file_path))
    tickers = lf_all.select("SYM_ROOT").unique().collect().get_column("SYM_ROOT").sort().to_list()

    print(f"📋 发现 {len(tickers)} 只股票，准备逐个计算...")

    all_summaries = []

    for ticker in tickers:
        # print(f"  -> 分析 {ticker} ...") # 如果股票太多可以注释掉这行

        # 2. 针对单个 Ticker 处理 (Lazy模式)
        # 过滤 -> 排序 -> 计算指标
        q = (
            lf_all
            .filter(pl.col("SYM_ROOT") == ticker)
            .sort("TS")
            .with_columns([
                # --- 核心指标计算 ---

                # SIDE: 简单的方向判断 (Price >= Mid 为买, 否则为卖)
                pl.when(pl.col("PRICE") >= pl.col("MID")).then(1).otherwise(-1).alias("SIDE"),

                # ARR_MID: 到达中间价 (这里假设当前行的 MID 就是到达中间价)
                pl.col("MID").alias("ARR_MID")
            ])
            .with_columns([
                # IS (Implementation Shortfall): (成交价 - 到达中间价) * 方向
                (10_000 * pl.col("SIDE") * (pl.col("PRICE") - pl.col("ARR_MID")) / pl.col("ARR_MID")).alias("IS_BPS"),

                # MI (Market Impact): (当前中间价 - 到达中间价) * 方向
                # 注意：如果 ARR_MID = MID，这一项会是 0。通常 MI 需要看未来 1秒或 5分钟的价格。
                (10_000 * pl.col("SIDE") * (pl.col("MID") - pl.col("ARR_MID")) / pl.col("ARR_MID")).alias("MI_BPS"),

                # MR 准备: 获取未来第 K 笔交易的中间价
                pl.col("MID").shift(-K_FWD_TRADES).alias("MID_FWD")
            ])
        )

        # 计算 MR (均值回归): (未来中间价 - 当前中间价) * -方向
        # 只有当未来有数据时才计算
        q = q.with_columns([
            pl.when(pl.col("MID_FWD").is_not_null())
            .then(10_000 * (-pl.col("SIDE")) * (pl.col("MID_FWD") - pl.col("MID")) / pl.col("MID"))
            .otherwise(None)
            .alias("MR_BPS")
        ])

        # 3. 聚合统计 (Aggregation)
        summary = q.select([
            pl.lit(ticker).alias("TICKER"),
            pl.len().alias("N_ROWS"),
            pl.col("SIZE").sum().alias("TOTAL_SIZE"),

            # VWAP
            (pl.col("PRICE") * pl.col("SIZE")).sum() / pl.col("SIZE").sum().alias("VWAP"),

            # 平均价差和深度
            (pl.col("SPREAD").mean() * 10_000).alias("AVG_SPREAD_BPS"),
            pl.col("DEPTH_TOB").mean().alias("AVG_DEPTH_TOB"),

            # IS 统计
            pl.col("IS_BPS").mean().alias("IS_MEAN_BPS"),
            pl.col("IS_BPS").median().alias("IS_MEDIAN_BPS"),
            pl.col("IS_BPS").min().alias("IS_MIN_BPS"),
            pl.col("IS_BPS").max().alias("IS_MAX_BPS"),

            # MI 统计
            pl.col("MI_BPS").mean().alias("MI_MEAN_BPS"),
            pl.col("MI_BPS").median().alias("MI_MEDIAN_BPS"),
            pl.col("MI_BPS").min().alias("MI_MIN_BPS"),
            pl.col("MI_BPS").max().alias("MI_MAX_BPS"),

            # MR 统计
            pl.col("MR_BPS").mean().alias("MR_MEAN_BPS"),
            pl.col("MR_BPS").median().alias("MR_MEDIAN_BPS"),
            pl.col("MR_BPS").min().alias("MR_MIN_BPS"),
            pl.col("MR_BPS").max().alias("MR_MAX_BPS"),

            pl.lit(K_FWD_TRADES).alias("FWD_TRADES_FOR_MR")
        ]).collect() # 执行计算

        all_summaries.append(summary)

    # 4. 合并所有结果
    if all_summaries:
        final_df = pl.concat(all_summaries)
        return final_df
    else:
        return None

# --- 执行并显示 ---
start_time = time.time()
summary_df = get_microstructure_stats(MERGED_FILE)

print(f"\n✅ 计算完成，耗时: {time.time() - start_time:.2f} 秒")
print("📊 统计摘要 (前 10 行):")

# 设置 Polars 显示选项以匹配截图宽度
pl.Config.set_tbl_cols(20)
pl.Config.set_tbl_width_chars(300)

display(summary_df)

# 如果需要保存结果：
# summary_df.write_csv("microstructure_summary.csv")
```
