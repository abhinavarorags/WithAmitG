```python
from typing import List

def solve(families: List[str], sizes: List[int]) -> str:
    # Work on a list of (family, size)
    arr = [(families[i], sizes[i]) for i in range(len(families))]

    while True:
        n = len(arr)
        # eater_of_victim[i] = index of eater that will eat i (or None)
        eater_of_victim = [None] * n
        # victim_of_eater[i] = index of victim that i will eat (or None)
        victim_of_eater = [None] * n

        fam = [f for f, _ in arr]
        sz = [s for _, s in arr]

        # One round: resolve from LEFT to RIGHT on potential victims
        for i in range(n):
            # If i already marked to be eaten, or is already eating someone,
            # it cannot be processed as a victim.
            if eater_of_victim[i] is not None:
                continue
            if victim_of_eater[i] is not None:
                continue

            chosen = None

            # Check LEFT neighbor as possible eater
            j = i - 1
            if j >= 0:
                if eater_of_victim[j] is None and victim_of_eater[j] is None:
                    if fam[j] != fam[i] and sz[j] > sz[i]:
                        chosen = j

            # If no left eater, check RIGHT neighbor
            k = i + 1
            if chosen is None and k < n:
                if eater_of_victim[k] is None and victim_of_eater[k] is None:
                    if fam[k] != fam[i] and sz[k] > sz[i]:
                        chosen = k

            # Register the eating event if any
            if chosen is not None:
                eater_of_victim[i] = chosen
                victim_of_eater[chosen] = i

        # If nobody is eaten this round, we reached equilibrium
        if not any(e is not None for e in eater_of_victim):
            break

        # Apply all eating events simultaneously
        new_arr = []
        for i in range(n):
            # Skip victims
            if eater_of_victim[i] is not None:
                continue

            f, s = arr[i]
            # If this one eats someone, add victim's size
            if victim_of_eater[i] is not None:
                s += arr[victim_of_eater[i]][1]
            new_arr.append((f, s))

        arr = new_arr

    # At the end, unique biggest microorganism
    best_family, best_size = max(arr, key=lambda x: x[1])
    return f"{best_family} {best_size}"


def main():
    l1 = ["A","B","C","D","D","C","C","A"]
    l2 = [20,20,20,20,4,4,30,2]

    ret = solve (l1,l2)
    print(ret)

main()
```
